```python
import serial.tools.list_ports
import time
import numpy as np
import cv2
from IPython.display import display, Image
from PIL import Image as PILImage
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.image import MIMEImage
from serial import Serial


# Load labels
labels = open('coco.names').read().strip().split('\n')
Name3E = ['dog', 'horse', 'elephant', 'cat', 'cow', 'sheep']
probability_minimum = 0.5
threshold = 0.3

# Load YOLOv3 model
net = cv2.dnn.readNetFromDarknet('yolov3.cfg', 'yolov3.weights')

# Open serial connection to Arduino
ser = Serial('COM8', 9600)  # Adjust the port accordingly

# Create a VideoCapture object with reduced frame size and resolution
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 320)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 240)

def send_email():
    sender_email = "sonali.subhasmita@btech.christuniversity.in"
    sender_password = "20759608"
    receiver_email = "sonalisubhasmita62@gmail.com"

    subject = "Motion Detection Alert!"
    
    if not confidences:
        print("No objects detected with sufficient confidence.")
        return

    # Find the object with the maximum confidence
    max_confidence_index = np.argmax(confidences)
    max_confidence_object = class_names[max_confidence_index]
    max_confidence = confidences[max_confidence_index]

    body = f"Detected object: {max_confidence_object} (Confidence: {max_confidence:.2f})"

    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = subject
    msg.attach(MIMEText(body, 'plain'))

    # Attach the image to the email
    with open("detected_image.png", "rb") as image_file:
        image = MIMEImage(image_file.read(), name="detected_image.png")
        msg.attach(image)

    # Connect to Gmail SMTP server
    with smtplib.SMTP('smtp.gmail.com', 587) as server:
        server.starttls()
        server.login(sender_email, sender_password)
        server.sendmail(sender_email, receiver_email, msg.as_string())


# Check if the camera is opened successfully
if not cap.isOpened():
    print("Error: Could not open camera.")
else:
    try:
        while True:
            if ser.in_waiting > 0:
                data = ser.readline().decode().strip()
                if data == "MotionDetected":
                    print("Motion detected!")

                    # Capture a frame from the webcam
                    ret, frame = cap.read()
                    
                    if ret is None or not ret:
                        print("Error: Failed to capture frame.")
                        continue

                    # Perform object detection on the captured frame
                    # Run inference through the network
                    height, width, channels = frame.shape
                    blob = cv2.dnn.blobFromImage(frame, 1/255.0, (416, 416), swapRB=True, crop=False)
                    net.setInput(blob)
                    output_from_network = net.forward(net.getUnconnectedOutLayersNames())

                    # Loop over each of the layer outputs
                    bounding_boxes = []
                    confidences = []
                    class_numbers = []
                    class_names = []
                    for output in output_from_network:
                        # Loop over each of the detections
                        for detection in output:
                            # Get the class ID and confidence of the detection
                            scores = detection[5:]
                            class_current = np.argmax(scores)
                            confidence_current = scores[class_current]

                            # Filter detections by confidence and class name
                            if confidence_current > probability_minimum and labels[class_current] in Name3E:
                                # Get the coordinates of the bounding box
                                box_current = detection[0:4] * np.array([width, height, width, height])
                                x_center, y_center, box_width, box_height = box_current.astype('int')

                                # Convert the center coordinates to top-left coordinates
                                x_min = int(x_center - (box_width / 2))
                                y_min = int(y_center - (box_height / 2))

                                # Add the bounding box to the list of bounding boxes
                                bounding_boxes.append([x_min, y_min, int(box_width), int(box_height)])
                                confidences.append(float(confidence_current))
                                class_numbers.append(class_current)
                                class_names.append(labels[class_current])

                    # Apply non-maximum suppression to remove overlapping bounding boxes
                    indexes = cv2.dnn.NMSBoxes(bounding_boxes, confidences, probability_minimum, threshold)

                    # Print detected animals and display the frame
                    for i in range(len(bounding_boxes)):
                        if i in indexes:
                            animal = class_names[i]
                            confidence = confidences[i]
                            print(f'Detected: {animal} (Confidence: {confidence:.2f})')

                            # Draw bounding box on the frame
                            x, y, w, h = bounding_boxes[i]
                            color = (0, 255, 0)
                            cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)

                    # Convert BGR to RGB for displaying with PIL
                    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

                    # Display the frame using IPython.display
                    display(Image(data=cv2.imencode('.png', frame_rgb)[1]))
                    cv2.imwrite("detected_image.png", frame)
                    send_email()

                    time.sleep(3)  # Add a delay to avoid continuous triggering
                    time.sleep(3)
                    cap.release()
                    break

    except KeyboardInterrupt:
        ser.close()
        cap.release()

```
